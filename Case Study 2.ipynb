```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
city = ['Bangalore','Mumbai','Gurgaon','Noida','New Delhi']
def f(s):
    if '/' in s:
        return s.split('/')[0].strip()
    return s.strip()

df = pd.read_csv('../Datasets/startup_funding.csv',index_col='SNo')
df.dropna(subset=['CityLocation'],inplace=True)
df['CityLocation'] = df['CityLocation'].apply(f)
df['CityLocation'] = df['CityLocation'].replace({
    'bangalore' : 'Bangalore',
    'Delhi' : 'New Delhi'
})
startups_location = df['CityLocation'].value_counts().loc[city].sort_values(ascending=False)

plt.bar(startups_location.index,startups_location)
plt.show()
```

```python
with open('../Datasets/startup_funding.csv') as file:
    file_data = csv.DictReader(file,skipinitialspace=True)
    file_list = list(file_data)
    
investor_name = []
for row in file_list:
    if ',' in row['InvestorsName']:
        names = row['InvestorsName'].split(',')
        for name in names:
            investor_name.append(name.strip())
    else:
        investor_name.append(row['InvestorsName'].strip())

investor_name = np.array(investor_name)
unique , count = np.unique(investor_name, return_counts=True)
d={}
for n in unique:
    d[n] = d.get(n,0) + count[n==unique].sum()
del d['']
d=sorted(d.items(),key=lambda x: x[1],reverse=True)[:5]
for k,v in d:
    print(k)
    
d = dict(d)
plt.pie( d.values() , labels= d.keys() , autopct='%.2f%%')
plt.show()
```

```python
startups = set()
investors = []
for row in file_list:
    investor = row['InvestorsName'].strip()
    startup = row['StartupName'].strip()
    if ',' in investor:
        for n in investor.split(','):
            if not((startup,n.strip()) in startups):
                investors.append(n.strip())
                startups.add((startup,n.strip()))
    else:
        if not((startup,investor) in startups):
            investors.append(investor.strip())
            startups.add((startup,investor.strip()))

x ,y = np.unique(investors,return_counts=True)

d={}

for i in x:
    d[i] = d.get(i,0) + y[i==x].sum()
del d['']
d=sorted(d.items(),key=lambda x: x[1],reverse=True)[:5]
for k,v in d:
    print(k)
d = dict(d)
plt.pie( d.values() , labels= d.keys() , autopct='%.2f%%')
plt.show()
```
